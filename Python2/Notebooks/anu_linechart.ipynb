```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('active_new_users.csv')
```

```python
df.head()
```

```python
df.date = pd.to_datetime(df.date)
plt.plot(df.date,df.count_new,label='new',linestyle='solid')
plt.plot(df.date,df.count_active,label='active',linestyle='solid')
plt.xlabel('date')
plt.title('active & new users')
plt.legend(loc='best')
plt.xticks(rotation=-60)
plt.show()
```
